```python
infected_data_frame = pd.read_csv(infected_dataset_url) #url yi oku ve DataFrame objesine ata
print(infected_data_frame.head())
```

```python
recovered_data_frame = pd.read_csv(recovered_dataset_url)
print(recovered_data_frame.head())
```

```python
deaths_data_frame = pd.read_csv(deaths_dataset_url)
print(deaths_data_frame.head())
```

```python
# total_infected_by_country = infected_data_frame.groupby('Country/Region').sum()
total_infected_by_country = infected_data_frame.groupby('Country/Region').sum().iloc[:, 3:].transpose()
print(total_infected_by_country)
```

```python
total_recovered_by_country = recovered_data_frame.groupby('Country/Region').sum().iloc[:, 3:].transpose()
print(total_recovered_by_country)
```

```python
total_deaths_by_country = deaths_data_frame.groupby('Country/Region').sum().iloc[:, 3:].transpose()
print(total_deaths_by_country)
```

```python
#Veri seti birleştirildi concat ile
birlestirilmis_data_frame = pd.concat([infected_data_frame, recovered_data_frame, deaths_data_frame], keys=["Infected", "Recovered", "Deaths"])
print(birlestirilmis_data_frame)
```

```python
# Province/State sütunundaki None değerler fillna ile dolduruldu
birlestirilmis_data_frame["Province/State"].fillna("Yok", inplace=True)# inplace false ise orjinal DataFrame değişmeden kalır ve değişiklikler yeni bir DataFrame'e uygulanır
print(birlestirilmis_data_frame)
```

```python
# Country/Region ve Province/State'e göre gruplanarak toplam vaka sayısı bulundu
toplam_vaka_sayisi = birlestirilmis_data_frame.groupby(["Country/Region", "Province/State"]).sum()
print(toplam_vaka_sayisi)
```

```python
sorted_total_cases = toplam_vaka_sayisi.sort_values(by="Country/Region", ascending=False)
print(sorted_total_cases)
```
