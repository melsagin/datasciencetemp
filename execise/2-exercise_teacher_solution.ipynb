```python
def make_frame(country:str):
    """Take country name as parameter, return infected, recovered, deaths values as timeseries table.
    
    """

    df = pd.DataFrame({
        'infected' : df_infected.loc[country],
        'recovered': df_recovered.loc[country],
        'deaths': df_deaths.loc[country]
    })
    df.index = pd.to_datetime(df.index, errors='coerce')
    # df.index = pd.to_datetime(df.index, format='%m/%d/%y')
    return df
```

##### VarsayalÄ±m ki new_infected adlÄ± bir sÃ¼tun, gÃ¼nlÃ¼k yeni enfekte olan sayÄ±larÄ± iÃ§eriyor. rolling(window=7).mean() bu sÃ¼tunu alÄ±r, her gÃ¼n iÃ§in bir pencere oluÅŸturur (bu pencere 7 gÃ¼nÃ¼ temsil eder) ve penceredeki 7 gÃ¼nÃ¼n ortalamasÄ±nÄ± alarak her bir gÃ¼n iÃ§in bir deÄŸer Ã¼retir. BÃ¶ylece, gÃ¼nlÃ¼k yeni enfekte olan sayÄ±larÄ±n daha dÃ¼zgÃ¼n bir eÄŸilimini gÃ¶steren bir grafik elde edebilirsiniz.

```python
df_us['new_infected_avg'] = df_us['new_infected'].rolling(window=7).mean()
# rolling fonksiyonu, bir zaman serisi Ã¼zerinde bir pencere (window) boyunca bir iÅŸlem gerÃ§ekleÅŸtirmek iÃ§in kullanÄ±lÄ±r
# rolling(window=7).mean() ifadesi, bir 7 gÃ¼nlÃ¼k hareketli ortalama hesaplayarak her bir gÃ¼n iÃ§in yeni bir deÄŸer oluÅŸturacaktÄ±r.
```

```python
df_us['new_infected_avg'].plot()
plt.show()
```

##### apply(lambda x: x[4:].sum() / x[:4].sum()): apply fonksiyonu, belirtilen fonksiyonu her pencere Ã¼zerinde Ã§alÄ±ÅŸtÄ±rmak iÃ§in kullanÄ±lÄ±r. Bu durumda, lambda ifadesi bir fonksiyonu temsil eder. Lambda fonksiyonu, her pencerenin son 4 deÄŸerini (x[4:]) toplar, ardÄ±ndan ilk 4 deÄŸeri (x[:4]) toplar ve bu iki toplamÄ±n oranÄ±nÄ± hesaplar. Bu, belirli bir penceredeki yeni enfekte olan sayÄ±larÄ±n belirli bir dÃ¶nemdeki deÄŸiÅŸim oranÄ±nÄ± hesaplamak iÃ§in kullanÄ±lÄ±r.

```python
# The basic reproduction number
df_us['Rt'] = df_us['new_infected'].rolling(8).apply(lambda x: x[4:].sum() / x[:4].sum())
```
