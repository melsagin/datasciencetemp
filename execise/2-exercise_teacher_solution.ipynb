```python
def make_frame(country:str):
    """Take country name as parameter, return infected, recovered, deaths values as timeseries table.
    
    """

    df = pd.DataFrame({
        'infected' : df_infected.loc[country],
        'recovered': df_recovered.loc[country],
        'deaths': df_deaths.loc[country]
    })
    df.index = pd.to_datetime(df.index, errors='coerce')
    # df.index = pd.to_datetime(df.index, format='%m/%d/%y')
    return df
```

##### Varsayalım ki new_infected adlı bir sütun, günlük yeni enfekte olan sayıları içeriyor. rolling(window=7).mean() bu sütunu alır, her gün için bir pencere oluşturur (bu pencere 7 günü temsil eder) ve penceredeki 7 günün ortalamasını alarak her bir gün için bir değer üretir. Böylece, günlük yeni enfekte olan sayıların daha düzgün bir eğilimini gösteren bir grafik elde edebilirsiniz.

```python
df_us['new_infected_avg'] = df_us['new_infected'].rolling(window=7).mean()
# rolling fonksiyonu, bir zaman serisi üzerinde bir pencere (window) boyunca bir işlem gerçekleştirmek için kullanılır
# rolling(window=7).mean() ifadesi, bir 7 günlük hareketli ortalama hesaplayarak her bir gün için yeni bir değer oluşturacaktır.
```

```python
df_us['new_infected_avg'].plot()
plt.show()
```

##### apply(lambda x: x[4:].sum() / x[:4].sum()): apply fonksiyonu, belirtilen fonksiyonu her pencere üzerinde çalıştırmak için kullanılır. Bu durumda, lambda ifadesi bir fonksiyonu temsil eder. Lambda fonksiyonu, her pencerenin son 4 değerini (x[4:]) toplar, ardından ilk 4 değeri (x[:4]) toplar ve bu iki toplamın oranını hesaplar. Bu, belirli bir penceredeki yeni enfekte olan sayıların belirli bir dönemdeki değişim oranını hesaplamak için kullanılır.

```python
# The basic reproduction number
df_us['Rt'] = df_us['new_infected'].rolling(8).apply(lambda x: x[4:].sum() / x[:4].sum())
```
